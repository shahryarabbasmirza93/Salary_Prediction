```python
import pandas as pd
import xml.etree.cElementTree as ET
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
Mdata=pd.read_csv('newtest2.csv')
```

```python
Mdata.head(5)
```

```python

Mdata=Mdata.drop(columns= ["Unnamed: 0"]) #Removing double index
```

```python
Mdata2= Mdata.drop(columns= ["ContractTime","Location"]) 

#A trial run to check which gives the best results,for this one we only consider Title and ContractTime
```

```python
df_3= pd.get_dummies(Mdata2,drop_first=True) # ONe hot encoding to convert categorical attributes to numerical attributes
```

```python
df_2 = pd.get_dummies(Mdata,drop_first=True) #Trail run for all four columns 
```

```python
df_3.head(3) 
```

```python
y= df_3.Salary
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_3, y, test_size=0.2)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
knn = KNeighborsClassifier()

knn_params = {
    'n_neighbors':list(range(1,100)),
    'weights':['distance','uniform']
}
knn_gs = GridSearchCV(knn, knn_params, cv=10)    #GridSearch for best model parameters
knn_gs.fit(X_train, y_train)
```

```python
knn_best = knn_gs.best_estimator_
print(knn_gs.best_params_)
print(knn_gs.best_score_)
```

```python
lr =  LogisticRegression()

lr_params = {
    'penalty':['l1','l2'],
    'C':np.logspace(-4, 2, 40),
    'solver':['liblinear']
}

lr_gs = GridSearchCV(lr, lr_params, cv=10, verbose=1)
lr_gs.fit(X_train, y_train)
```

```python
best_lr = lr_gs.best_estimator_
print(lr_gs.best_params_)
print(lr_gs.best_score_)
```

```python
RFC =  RandomForestClassifier()

rfc_params = {
 'bootstrap': [True, False],
 'max_depth': [None, 1, 2, 3],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [25, 50, 100]
}

rfc_gs = GridSearchCV(RFC, rfc_params, cv=10, verbose=1)
rfc_gs.fit(X_train, y_train)
```

```python
rfc_best = rfc_gs.best_estimator_
print((rfc_gs.best_params_))
print((rfc_gs.best_score_))
```

```python
evaluation = pd.DataFrame()
evaluation['Classifier'] = ['KNN', 'Logistic Regression', 'Random Forest']
evaluation['Training Accuracy'] = [round(knn_gs.best_score_, 3),
                                             round(lr_gs.best_score_, 3),
                                             round(rfc_gs.best_score_, 3)] 
evaluation['Testing Accuracy'] = [round(knn_best.score(X_test, y_test), 3),
                                  round(best_lr.score(X_test, y_test), 3),
                                 round(rfc_best.score(X_test, y_test), 3)] 
evaluation.sort_values(by='Testing Accuracy', ascending=False)
```
